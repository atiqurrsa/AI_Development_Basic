```python
#defult x points
#uses of marker (o,*)

y=([2,6,3,9,3,5,8,2,6])
plt.plot(y,marker='o')
plt.show()
```

```python
# FMT prameter (marker,line,color)

y=([2,5,1,7,2,9,4,9])
plt.plot(y,'o:r')
plt.show()
```

```python
#marker size(ms) & marker edge color (mec) & marker face color (mfc) 
#line width (lw)

y=([2,5,1,7,2,9,4,9])
plt.plot(y,'o:g',ms=20,mec='r',mfc='c',lw=5)
plt.show()
```

```python
# Titel & titel position
# Labels
# Grid method

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([200, 290, 220, 300, 280, 230, 300, 360, 320, 330])

plt.title("Sports Watch Data",loc='right',c='m')
plt.xlabel("Average Pulse",c='r')
plt.ylabel("Clorie Burnage",c='b')


plt.plot(x,y)
plt.grid()
plt.show()
```

```python
#grid axis,color,linestyle,linewidth

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([200, 290, 220, 300, 280, 230, 300, 360, 320, 330])

plt.title("Sports Watch Data",loc='right',c='m')
plt.xlabel("Average Pulse",c='r')
plt.ylabel("Clorie Burnage",c='b')


plt.plot(x,y)
plt.grid(axis='y',color='g',linestyle=':',linewidth='1')
plt.show()
```

```python
#subplot with 3 row and 2 column

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(3,2,1)
plt.plot(x,y)

x = np.array([0, 4, 10, 9])
y = np.array([10, 2, 1, 10])

plt.subplot(3,2,2)
plt.plot(x,y)

x = np.array([0, 4, 2, 9])
y = np.array([3, 2, 1, 10])

plt.subplot(3,2,3)
plt.plot(x,y)

x = np.array([0, 4, 2, 9])
y = np.array([3, 2, 5, 10])

plt.subplot(3,2,4)
plt.plot(x,y)


x = np.array([0, 4, 2, 9])
y = np.array([3, 9, 1, 10])

plt.subplot(3,2,5)
plt.plot(x,y)

x = np.array([9, 4, 2, 9])
y = np.array([3, 2, 1, 10])

plt.subplot(3,2,6)
plt.plot(x,y)



plt.show()
```

```python
#scatter plot

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x,y,color='green')

x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])

plt.scatter(x,y,color='red')

plt.xlabel("AGE")
plt.ylabel("CAR SPEED")

plt.colorbar()
plt.show()
```

```python
#bar

x = np.array(["A", "B", "C", "D","E", "H","I"])
y = np.array([3, 8, 1, 10,12, 33,12])

plt.bar(x,y,width=0.3)
plt.show()
```

```python
#histogram with random data 

import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(170, 10, 250)

plt.hist(x)
plt.show() 
```

```python
#pie chart
y=np.array([3,6,4,9,8])
mylabel=["apple","banana","am","kathal","jam"]
myexplode=[0.1,0,0.2,0,0]
mycolor=["red","green","blue","black","pink"]
plt.pie(y,labels=mylabel,explode=myexplode,colors=mycolor,shadow=True)
plt.legend(title="Fruits",loc="upper right")
plt.show()
```
